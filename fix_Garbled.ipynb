# CoraboのRで日本語文字化け解決方法

```
install.packages("showtext")
library(showtext)
# デフォルトで使えるフォント一覧の表示
font_files()
# Google fontをインストール
# 字体の確認はこちら: https://fonts.google.com/?subset=japanese
# "Noto Sans JP", "Noto Serif JP", "M PLUS 1p", "M PLUS Rounded 1c", 
# "Sawarabi Mincho", "Sawarabi Gothic", "Kosugi Maru", "Kosugi"
# の中からどれか選ぶ。 複数使いたい場合は、font_add_google()を繰り返す。
font_add_google("Sawarabi mincho")
# 追加したフォントの有効化
showtext_auto()
par(family= "Sawarabi mincho") #フォント指定-> 文字化け解消！
boxplot(iris$Petal.Length ~ iris$Species,
        col = "orange",
        horizontal = T,
        main = "アヤメの花弁の長さの分布")
# ggplot2ライブラリを使って描く
library(ggplot2)
ggplot(iris) +                           #データフレームの指定
  geom_boxplot(aes(y = Petal.Length,        　#縦軸の変数
                   x = Species,       #グループ化を行う変数
                   fill=Species,     　#塗り分けの対象となる変数
                   alpha=0.7)) +    #透明度の指定
  labs(title = "あやめの花弁の長さ") + 
  theme_bw(base_family = "Sawarabi mincho")
```
