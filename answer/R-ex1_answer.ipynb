**Step1**　必要なライブラリを読み込む

```python
library(ggplot2)
```

**Step2**　CSVファイルを読み込んでデータフレームに格納

```python
DF <- read.table("https://raw.githubusercontent.com/mkato-riken/test/main/sample.csv", sep =",", header = T)
```

**Step3**　先頭の６行を確認

```python
head(DF)
```

**Step４**　行数、列数を確認

```python
str(DF)
```

**Step5**　平均値、最大値、最小値などを確認

```python
summary(DF)
```

**Step6**　varAとvarBをそれぞれ２乗して足した値が１未満かどうかを判定（判定結果を id　に入れる）

```python
id <- DF$varA^2 + DF$varB^2 < 1
```

**Step7**　IDという名称の列を作り "steelblue"という色の名称を入れる

```python
DF$ID <- "steelblue"
```

**Step8**　idの値がTRUEであるケース（行）だけ、IDの値を"darkblue"に置き換える

```python
DF$ID[id==T] <- "darkblue"
```

**Step9**　IDの値について頻度を計算

```python
table(DF$ID)
```

**Step10**　散布図を描く（横軸に用いる変数：varA　縦軸に用いる変数：varB　色の名称：ID)

```python
ggplot() +
geom_point(aes(DF$varA, DF$varB),
            color=DF$ID, alpha=0.6, size=4)
```
